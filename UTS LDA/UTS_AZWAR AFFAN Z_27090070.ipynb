```python
print("=== PROGRAM PERHITUNGAN GAJI KARYAWAN ===")

karyawan =  [
    {"nama": "Andi", "gaji_pokok": 4500000, "lama_kerja": 2},
    {"nama": "Budi", "gaji_pokok": 5000000, "lama_kerja": 5},
    {"nama": "Citra", "gaji_pokok": 6500000, "lama_kerja": 8}
]

print("\n=== TOTAL GAJI KARYAWAN ===")

for x in karyawan:
    nama = x["nama"]
    gaji_pokok = x["gaji_pokok"]
    lama_kerja = x["lama_kerja"]
    
    if lama_kerja >= 5:
        bonus = gaji_pokok * 0.15
    elif lama_kerja >= 3:
        bonus = gaji_pokok * 0.10
    else:
        bonus = gaji_pokok * 0.05
    
    total_gaji = gaji_pokok + bonus
    
    print(f"{nama}: Rp {total_gaji:,.0f}")
```
